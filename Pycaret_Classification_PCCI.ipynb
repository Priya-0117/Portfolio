```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/nvamsimohan/DallasDSA/main/PCCI%20Cases%20Readmit30%20Data.csv')

df

```

```python
df.shape
```

```python
# target variable
y = 'Readmit30'
#setting ignored variables
ignore_vars = ['Weight', 'Height']
```

```python
# prompt: proportion  and count of people that are readmited?

df[y].value_counts()
```

```python
#setting categorical variables
cat_vars = cat_vars = ['Admit_Week', 'Admit_Month', 'Gender', 'Marital_Status',
                      'Insurance_Provider', 'Tobacco_User', 'Depression', 'ICU', 'Drug_Abuse',
                      'Mood_Disorder', 'Diabetes', 'Anxiety', 'Obesity', 'Dementia',
                      'Condition',
                      'Care_Plan_Following_Discharge']

#setting numerical varibles
num_vars = ['Age','Bmi', 'Pulse', 'Temperature', 'Pat_Pain_Score',
       'ER_Visits', 'IP_Visits', 'Chronic_Conditions', 'Glucose', 'Cost_Of_Initial_Stay']
```

```python
from pycaret.classification import *
```

```python
# Configuration of model
classification_model = setup(data = df, target = y,
                                 ignore_features = ignore_vars,
                                 categorical_features = cat_vars,
                                 numeric_features = num_vars, train_size=0.8)
```

```python
compare_models(budget_time= 100)
```

```python
#building model using the RFC
best_model = create_model('rf')
```

```python
# Display confusion metrics
plot_model(best_model, plot = 'confusion_matrix')
```

```python
evaluate_model(best_model)
```

```python
#Predicting Readmit status for the client data patients
client_data = pd.read_csv('https://raw.githubusercontent.com/nvamsimohan/DallasDSA/main/PCCI%20-%20Client%20Data%20File.csv')
readmit_status_prediction = predict_model(best_model, data = client_data)

readmit_status_prediction
```

````python
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/nvamsimohan/DallasDSA/main/Zillow%20Tarrant%20County%20Raw%20Data.csv')

data
````

````python
data.columns
````

````python
from pycaret.anomaly import *
````

````python
anomaly_model = setup(data)
````

````python
#builing an anomaly model using knn alg
knn_model = create_model('knn')
````

````python
# identifying the anomalies
knn_anomalies = predict_model(knn_model,data=data)
knn_anomalies
````

````python
#exporting Anomalies to csv file
knn_anomalies.to_csv('knn_anomalies.csv')
````

````python
# count the number of anomalies
knn_anomalies['Anomaly'].value_counts()

#anomaly score helps us cross check the threshold
````

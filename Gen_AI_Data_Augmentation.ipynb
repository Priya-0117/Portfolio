```python
# Cs 2
# Importing the dataset
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/nvamsimohan/DallasDSA/main/Program%20Management%20Data.csv')

data
```

```python
# cs 3
import sdv
```

```python
# Cs 4
# Importing SIngle table data augmentation modules
from sdv.metadata import SingleTableMetadata
metadata = SingleTableMetadata()
```

```python
# cs 5
# Deriving metadata from the table
metadata.detect_from_dataframe(data)
```

```python
# Cs 6
# metadata validation
metadata.validate()

metadata.visualize()
```

```python
# cs 7
# Importing synthetic data generator algorithm
from sdv.single_table import GaussianCopulaSynthesizer

synthesizer = GaussianCopulaSynthesizer(metadata)
```

```python
# cs 8
# Train the synthesizer
synthesizer.fit(data)
```

```python
# Cs 9
# Generating synthetic data
synthetic_data = synthesizer.sample(num_rows=100)

synthetic_data
```

```python
# Cs 10
# Evaluating data quality
from sdv.evaluation.single_table import evaluate_quality

quality_report = evaluate_quality(real_data=data, synthetic_data=synthetic_data, metadata=metadata)

quality_report
```
